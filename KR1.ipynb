```python
import librosa
import numpy as np
```

```python
# Загрузка звуковых дорожек
wet_1_data, wet_1_rate = librosa.load('genres/wet/wet.wav', sr=None)
wet_2_data, wet_2_rate = librosa.load('genres/wet/wet2.wav', sr=None)
dry_data, dry_rate = librosa.load('genres/dry/dry.wav', sr=None)
static_data, static_rate = librosa.load('genres/static/static.wav', sr=None)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
plt.figure(figsize=(14, 5))
librosa.display.waveplot(wet_1_data, sr=wet_1_rate)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
plt.figure(figsize=(14, 5))
librosa.display.waveplot(wet_2_data, sr=wet_2_rate)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
plt.figure(figsize=(14, 5))
librosa.display.waveplot(dry_data, sr=dry_rate)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
plt.figure(figsize=(14, 5))
librosa.display.waveplot(static_data, sr=static_rate)
```

```python
wet_1_spec = librosa.feature.melspectrogram(y=wet_1_data, sr=wet_1_rate)
wet_2_spec = librosa.feature.melspectrogram(y=wet_2_data, sr=wet_2_rate)
dry_spec = librosa.feature.melspectrogram(y=dry_data, sr=dry_rate)
static_spec = librosa.feature.melspectrogram(y=static_data, sr=static_rate)
```

```python
# Вывод спектрограмм на экран
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(wet_1_spec, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Wet 1 Spectrogram')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(wet_2_spec, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Wet 2 Spectrogram')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(dry_spec, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Dry Spectrogram')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(static_spec, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Static Spectrogram')
plt.tight_layout()
plt.show()
```
